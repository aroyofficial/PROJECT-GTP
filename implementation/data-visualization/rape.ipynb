```python
# import the dataset for 2016 rape victims
import pandas as pd
df1 = pd.read_csv('2016 Victims of Rape.csv')
df1.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
age_dict = {
    'Child Victims of Rape (Below 18 Yrs) - Below 6 Years': '<6',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above - Below 12 Years': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above - Below 16 Years': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above - Below 18 Years': '>=16-<18',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl /Child Victims': 'Total Child Victims',
    'Women Victims of Rape (Above 18 Yrs) - 18 Years & Above - Below 30 Years': '>=18-<30',
    'Women Victims of Rape (Above 18 Yrs) - 30 Years & Above - Below 45 Years': '>=30-<45',
    'Women Victims of Rape (Above 18 Yrs) - 45 Years & Above - Below 60 Years': '>=45-<60',
    'Women Victims of Rape (Above 18 Yrs) - 60 Years & Above': '>=60',
    'Women Victims of Rape (Above 18 Yrs) - Total Women/Adult Victims': 'Total Adult Victims',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above': '>=16-<18',
    'State/UT': 'Area'
}
df1.rename(columns = age_dict, inplace = True)
df1.drop([df1.columns[0]], axis = 1, inplace = True)
df1
```

```python
# import the dataset for 2017 rape victims
df2 = pd.read_csv('2017 Victims of Rape.csv')
df2.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
df2.rename(columns = age_dict, inplace = True)
df2.drop([df2.columns[0]], axis = 1, inplace = True)
df2
```

```python
# import the dataset for 2018 rape victims
df3 = pd.read_csv('2018 Victims of Rape.csv')
df3.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
age = {
    'Child Victims of Rape (Below 18 Yrs) - Below 6 Years': '<6',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl /Child Victims': 'Total Child Victims',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above ': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above ': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above ': '>=16-<18',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl/Child Victims': 'Total Child Victims',
    'Women Victims of Rape (Above 18 Yrs) - 18 Years & Above - Below 30 Years': '>=18-<30',
    'Women Victims of Rape (Above 18 Yrs) - 30 Years & Above - Below 45 Years': '>=30-<45',
    'Women Victims of Rape (Above 18 Yrs) - 45 Years & Above - Below 60 Years': '>=45-<60',
    'Women Victims of Rape (Above 18 Yrs) - 60 Years & Above': '>=60',
    'Women Victims of Rape (Above 18 Yrs) - Total Women/Adult Victims': 'Total Adult Victims',
    'State/UT ': 'Area'
}
df3.rename(columns = age, inplace = True)
df3.drop([df3.columns[0]], axis = 1, inplace = True)
df3
```

```python
import matplotlib.pyplot as plt
import numpy as np
def analysis(df, s):
    x = np.arange(len(df))
    w = 0.2
    plt.figure(figsize = (12, 8))
    plt.bar(x, df.iloc[:, 1], w)
    plt.bar(x + w,  df.iloc[:, 2], w)
    plt.bar(x + 2 * w,  df.iloc[:, 3], w)
    plt.bar(x + 3 * w,  df.iloc[:, 4], w)
    plt.bar(x + 4 * w,  df.iloc[:, 6], w)
    plt.bar(x + 5 * w,  df.iloc[:, 7], w)
    plt.bar(x + 6 * w,  df.iloc[:, 8], w)
    plt.bar(x + 7 * w,  df.iloc[:, 9], w)
    plt.xticks(x, df.iloc[:, 0], rotation = 90)
    s = s + ' Rape Victims Analysis'
    plt.title(s)
    plt.ylabel('No. of Victims')
    plt.legend(['<6', '>=6-<12', '>=12-<16', '>=16-<18', '>=18-<30', '>=30-<45', '>=45-<60', '>=60'])
    plt.show()
```

```python
analysis(df1, '2016')
analysis(df2, '2017')
analysis(df3, '2018')
```

```python
def total_victims(df, s):
    x = np.arange(len(df))
    w = 0.5
    plt.figure(figsize = (12, 8))
    plt.bar(x, df.iloc[:, 5], w)
    plt.bar(x + w,  df.iloc[:, 10], w)
    plt.bar(x + (2 * w),  df.iloc[:, 11], w)
    plt.ylabel('No. of Victims')
    s = s + ' Total Rape Victims Statewise'
    plt.title(s)
    plt.xticks(x, df.iloc[:, 0], rotation = 90)
    plt.legend(['Total Child Victims', 'Total Adult Victims', 'Total Victims'])
    plt.show()
```

```python
total_victims(df1, '2016')
total_victims(df2, '2017')
total_victims(df3, '2018')
```
